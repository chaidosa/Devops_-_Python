#**Pyplot Tutorial:**

**matplotlic pyplot** is a collection of command style functions that make matplotlib work like MATLAB.

Each pyplot function makes some changes to figure: e.g, creates a figure, creates a plotting area in a figure, plot some lines in a plotting area, decorates the plot with labels, etc.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2, 4, 6, 4])
plt.ylabel("Numbers")
plt.xlabel('Indices')
plt.title('MyPlot')
plt.show()
```

```python
#Plot X versus Y

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('square')
plt.xlabel('numbers')
plt.grid() #grid on

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') #ro r-color o-> means put o shape dots
plt.ylabel('square')
plt.xlabel('numbers')
plt.grid() #grid on

plt.show()
```

If matplotlib were limited to working with lists, it would be faurly useless for numeric processing. Generally, you use **numpy arrays**. In fact, all sequences are converted to numpy arrays internally.


```python
import numpy as np
```

```python
t = np.arange(0., 5., 0.2)
print(t)
```

```python
#bule dashes, red squares and green triangles
plt.plot(t, t**2, 'b--', label = '^2')
plt.plot(t, t**2.2, 'rs', label = '^2.2')
plt.plot(t, t**2.5, 'g^', label = '^2.5')
plt.grid()
plt.legend() # add legend based on line labels
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, linewidth = 3.5)
plt.show()
```

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [2, 4, 6, 8]
lines = plt.plot(x1, y1, x2, y2)

#use keywords args
plt.setp(lines[0], color = 'r', linewidth = 2.0)

# or MATLAB style string value pairs
plt.setp(lines[1], 'color', 'g', 'linewidth', 2.0)
plt.grid()
```

#**Working with multiple figures and axes**

```python
def f(t):
  return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.2)

plt.figure(1) #create a single fugure
#The sublplot() command specifies numrows, numcols,
# fignum where fignum ranges from 1 to numrows*numcols.
plt.subplot(211) #no of row, no of col, fig number here 2->numrow, 1->numcol, 1->fignumber
plt.grid()
plt.plot(t1, f(t1), 'b--')

plt.subplot(212)  #no of row, no of col, fig number here 2->numrow, 1->numcol, 2->fignumber
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure(1)     #the first figure
plt.subplot(211)  # the fist subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)  # the second subplot
plt.plot([4, 5, 6])


plt.figure(2) # second figure
plt.plot([4, 5, 6]) #creates a subplot(111) by default

 #go back to figure 1
plt.figure(1) #figure 1 current; subplot(212) still 
plt.subplot(211)
plt.title('Easy as 1, 2, 3') #subplot 211 title
plt.show()
```
