```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = pd.read_csv("iris.csv")
```

```python
iris
```

```python
import numpy as np
iris_setosa = iris.loc[iris['variety'] == 'Setosa'];
iris_virginica = iris.loc[iris['variety'] == 'Virginica'];
iris_versicolor = iris.loc[iris['variety'] == 'Versicolor'];
```

```python
#mean, variance , std-deviation
print("Means: ")
print(np.mean(iris_setosa["petal.length"]))

print(np.mean(iris_virginica['petal.length']))
print(np.mean(iris_versicolor['petal.length']))

#mean with an outlier. if someone entered wrong data-> such a point is 
# often called outlier.
print(np.mean(np.append(iris_setosa["petal.length"], 50)))

#Frome here we understood that a slight change in measurement or
# mistakenly entering wrong data can cause lot of problems. and the
# mean would jump drastically and we won't be able to make sense of
# the data using mean

print("\nStd-dev:")
print(np.std(iris_setosa['petal.length']))
print(np.std(iris_virginica['petal.length']))
print(np.std(iris_versicolor['petal.length']))
```

```python
print(np.var(iris_setosa['petal.length']))
```

# Median, Percentile, Quantile, IQR, MAD

Advantage of median is, by introducing outlier doesn't effect much on
our data our data changes when more than half of the values gets corrupted.

```python
print('\nMedians:\niris_setosa')
print(np.median(iris_setosa['petal.length']))

print("\nMeadian with outlier")
print(np.median(np.append(iris_setosa['petal.length'], 50)))

print("\nMedian of iris virginica")
print(np.median(iris_virginica['petal.length']))
print("\nMedian of iris versicolor")
print(np.median(iris_versicolor['petal.length']))
```

```python
print('\nQuantiles"')
print(np.percentile(iris_setosa['petal.length'], np.arange(0, 100, 25)))
print('\n 25, 50 , 75, 100 percentiles\n')
print(np.percentile(iris_setosa['petal.length'], np.array([25, 50, 75, 100])))
print('\n')
print(np.percentile(iris_virginica['petal.length'], np.array([25, 50, 75, 100])))
print('\n')
print(np.percentile(iris_versicolor['petal.length'], np.array([25, 50, 75, 100])))
print('\n')
print('90th percentile\n')
print(np.percentile(iris_versicolor['petal.length'], np.array([90])))
```

```python
from statsmodels import robust
print("\n Median Absolute Deviation:")
print(robust.mad(iris_setosa['petal.length']))
print(robust.mad(iris_virginica['petal.length']))
print(robust.mad(iris_versicolor['petal.length']))
```

# Box plot and Whiskers

```python
#Box plot with whiskers: another method of visualizing the 1-D scatter plot
# The concept of median, percentile, quantile.
# How to draw the box in the box-plot?
# How to draw whiskers:[no standard way] could use min and max or use
# other IQR like idea.

#Note: In the plot below, a technique call inter-quartile range is used
# in whiskers in the plot below donot correspond to the min and max
# values.


# Box-plot can be visualized as a PDF on the sideways

sns.set_style('whitegrid')
sns.boxplot(x = 'variety', y = 'petal.length', data = iris)
plt.show()
```

# Violin plots

A violin plot combines the benifits of the previous two plots(histogram and boxplot) and simplifies them.

Denser regions of the data are fatter, and sparser ones thinner in a 
violin plot

```python
sns.set_style('whitegrid')
sns.violinplot(x='variety', y = 'petal.length', data = iris, size = 8)
plt.show()
```

# Multivariate probablity density, contour plot  

```python
#2D density plot, contors-plot
sns.jointplot(x='petal.length', y = 'petal.width', data = iris, kind = 'kde')
plt.show()
```
