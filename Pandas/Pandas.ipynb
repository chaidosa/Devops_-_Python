#**Introduction to Pandas Dataframe**

Dataframe is a main object in pandas. it used to represent data with rows and columns.

Data frame is a datastructure represent the data in tabular or excel spread sheet like data

```python
import pandas as pd

df = pd.read_csv('weather_data.csv')
df
```

```python
#list of tuples -> another way to form dataframe

weather_data = [('1/1/2017', 32, 6, 'Rain'),
                ('1/2/2017', 35, 7, 'Sunny'),
                ('1/3/2017', 28, 2, 'Snow'),
                ('1/4/2017', 24, 7, 'Snow')]

df = pd.DataFrame(weather_data, columns=['day', 'temp', 'windSpeed', 'dataLike'])
df
```
