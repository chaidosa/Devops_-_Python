# Haberman Breast Cancer Survivial Dataset

This dataset describes breast cancer patient data and the outcome is the patient survival.

**AGE** It reprsents the age of the patient at which they undergone surgery (age from 30 to 83)<br>
**Operation year** Year in which patient was undergone sugery(1958-1969)<br>
**Axillary nodes(Lymph Nodes)** Lymph nodes are small, bean-shaped organs that acts as filters along the lymph fluid channels. As lymph fluid leaves the breast and eventually goes back into the blodstream, the lymph nodes try to catch and trap cancer cells before they reach other parts of the body. Having cancer cells in the lymph nodes under your arm suggest an increased risk of the cancer spreading. Ind our data it is axillary nodes detected (0-52).<br>
**Survival Status** It represents whether patient survive more than 5 years or less after undergone through surgery. Here if patients survived 5 years or more is represented as 1 and patients who survived less than 5 years is represented as 2. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
source = 'haberman.csv'
names = ['Age', 'Year operation', 'Axillary nodes', 'Survival status']
df = pd.read_csv(source, names = names)
```

```python
df.plot(kind = 'scatter', x = 'Axillary nodes', y = 'Age')
plt.grid()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(df, hue = 'Survival status', height = 4) \
.map(plt.scatter, 'Age', 'Axillary nodes').add_legend()
```

```python
plt.close()
sns.set_style('whitegrid')
sns.pairplot(df, hue = "Survival status", height = 3, vars = ['Age', 'Year operation', 'Axillary nodes'])
plt.show()
```
