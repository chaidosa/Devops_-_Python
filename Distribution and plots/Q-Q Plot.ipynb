```python
import numpy as np
import pylab
# library used for statistical tools
import scipy.stats as stats

#N(0, 1) -> Gaussian distribution
std_normal = np.random.normal(loc = 0, scale = 1, size = 1000)

# 0 to 100 percentile-> basically its percentile of data
for i in range(0, 101):
    print(i, np.percentile(std_normal, i))
```

```python
import matplotlib.pyplot as plt

#generate 100 samples from N(20, 5)


#try changing the size
measurements = np.random.normal(loc = 0, scale = 1, size = 500)

plt.xlim(-1,1)

stats.probplot(measurements, dist = 'norm', plot = pylab)
pylab.show()
```

```python
#if we used different graphs 

measurements = np.random.uniform(low = 0, high = 4, size = 500)

plt.xlim(-4, 4)

stats.probplot(measurements, dist = 'norm', plot= pylab)
pylab.show()
```

**#to sum up all the idea here is to get two different measurements and 
find the percentiles of both data (both of the datas popultation can be different) and plot it, If all the points lie in a line or similar 
to a line then we can say that both of the graph follow same distributions.**
